```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('curahhujan_dataset.csv', sep=";")
```

```python
df['prediksi']=df.iloc[:,2].rolling(window=3).mean()
```

```python
df
```

```python
df['prediksi']=df['prediksi'].replace(np.nan,0)
df
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(df['prediksi'],df['jumlah_curah_hujan'])
rmse=np.sqrt(mse)
print("MSE: ",mse)
print("RMSE: ",rmse)
```

```python
a=range(len(df))
plt.scatter(a, df['jumlah_curah_hujan'], color = 'blue')
plt.plot(a, df['prediksi'], color = 'red')
plt.title('Curah Hujan di Jawa Barat')
plt.xlabel('Data ke-')
plt.ylabel('Curah Hujan')
plt.show()
```
